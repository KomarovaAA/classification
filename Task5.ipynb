```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
el_tab = pd.read_csv("el_table.csv").drop(columns="Unnamed: 0")
el_tab.head()
```

# Явка:

```python
el_tab["Сумма"] = el_tab["Число недействительных избирательных бюллетеней"] + el_tab["Число действительных избирательных бюллетеней"]
el_tab["Явка"] = el_tab["Сумма"]/el_tab["Число избирателей, внесенных в список избирателей на момент окончания голосования"]
el_tab.sort_values("Явка", ascending=False, inplace=True)
el_tab.head()
```

# Наибольший результат Тихоновой:

```python
el_tab["Процент проголосовавших за Тихонову"] = el_tab["Тихонова Надежда Геннадьевна"]/el_tab["Сумма"]*100
pd.set_option('mode.use_inf_as_na', True)
max_per = el_tab["Процент проголосовавших за Тихонову"].max(skipna=None)
print("Максимальный процент:", max_per)
el_tab.loc[el_tab["Процент проголосовавших за Тихонову"] == max_per]
```

# Тик, где разница между наибольшей явкой и наименьшей максимальна:

```python
dif_tik = (el_tab.groupby("Tik")["Явка"].max() - el_tab.groupby("Tik")["Явка"].min()).sort_values(ascending=False)
dif_tik[:1]
```

# Дисперсия явки:

```python
el_tab.groupby("Tik")["Явка"].var().sort_values()
```

# Проценты и уики:

```python
el_tab["Процент проголосовавших за Беглова"] = np.around(el_tab["Беглов Александр Дмитриевич"]/el_tab["Сумма"]*100)
el_tab["Процент проголосовавших за Амосова"] = np.around(el_tab["Амосов Михаил Иванович"]/el_tab["Сумма"]*100)
el_tab["Процент проголосовавших за Тихонову"] = np.around(el_tab["Тихонова Надежда Геннадьевна"]/el_tab["Сумма"]*100)
el_tab.head()
```

```python
#Тихонова
el_tab.groupby("Процент проголосовавших за Тихонову").count()["Uik"]
```

```python
#Беглов
el_tab.groupby("Процент проголосовавших за Беглова").count()["Uik"]
```

```python
#Амосов
el_tab.groupby("Процент проголосовавших за Амосова").count()["Uik"]
```

# Явка и результат:

```python
tih = el_tab.groupby("Явка")["Процент проголосовавших за Тихонову"].mean()
beg = el_tab.groupby("Явка")["Процент проголосовавших за Беглова"].mean()
amos = el_tab.groupby("Явка")["Процент проголосовавших за Амосова"].mean()
```

```python
#синий - Тихонова, розовый - Беглов, фиолетовый - Амосов
plt.figure(figsize=(16, 9))
plt.scatter(tih.index, tih.values, c="blue", s = 10, alpha=0.4)
plt.scatter(beg.index, beg.values, c="deeppink", s = 10, alpha=0.4)
plt.scatter(amos.index, amos.values, c="purple", s = 10, alpha=0.4)
plt.xlabel("Явка")
plt.ylabel("Проценты проголосовавших")
plt.show()
```
